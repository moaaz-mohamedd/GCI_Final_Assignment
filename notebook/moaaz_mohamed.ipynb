```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from pathlib import Path

pd.set_option("display.max_columns", 150)
pd.set_option("display.max_rows", 100)
pd.set_option("display.width", 200)

# Detect project directory
CURRENT_DIR = Path.cwd()

# لو النوتبوك شغال من فولدر notebook، اطلع خطوة لفوق
if CURRENT_DIR.name.lower() == "notebook":
    PROJECT_DIR = CURRENT_DIR.parent
else:
    PROJECT_DIR = CURRENT_DIR

DATA_DIR = PROJECT_DIR / "data"
OUTPUT_DIR = PROJECT_DIR / "outputs"
CHARTS_DIR = OUTPUT_DIR / "charts"
TABLES_DIR = OUTPUT_DIR / "tables"

# Create output folders if not exists
CHARTS_DIR.mkdir(parents=True, exist_ok=True)
TABLES_DIR.mkdir(parents=True, exist_ok=True)

print("Current directory:", CURRENT_DIR)
print("Project directory:", PROJECT_DIR)
print("Data directory:", DATA_DIR)
print("Charts directory:", CHARTS_DIR)
print("Tables directory:", TABLES_DIR)
```

```python
csv_files = list(DATA_DIR.glob("*.csv"))

print("CSV files found:")
for file in csv_files:
    print("-", file.name)

print("\nNumber of CSV files:", len(csv_files))
```

```python
client_path = DATA_DIR / "Client.csv"
record_path = DATA_DIR / "Record.csv"

client = pd.read_csv(client_path, low_memory=False)
record = pd.read_csv(record_path, low_memory=False)

print("Client shape:", client.shape)
print("Record shape:", record.shape)
```

```python
print("CLIENT SAMPLE")
display(client.head())

print("RECORD SAMPLE")
display(record.head())
```

```python
print("Client columns:")
print(client.columns.tolist())

print("\nRecord columns:")
print(record.columns.tolist())
```

```python
print("CLIENT INFO")
client.info()

print("\n" + "="*80 + "\n")

print("RECORD INFO")
record.info()
```

```python
print("CLIENT Customer_ID check")
print("Rows:", len(client))
print("Unique Customer_ID:", client["Customer_ID"].nunique())
print("Duplicated Customer_ID:", client["Customer_ID"].duplicated().sum())
print("Missing Customer_ID:", client["Customer_ID"].isna().sum())

print("\nRECORD Customer_ID check")
print("Rows:", len(record))
print("Unique Customer_ID:", record["Customer_ID"].nunique())
print("Duplicated Customer_ID:", record["Customer_ID"].duplicated().sum())
print("Missing Customer_ID:", record["Customer_ID"].isna().sum())
```

```python
client_ids = set(client["Customer_ID"])
record_ids = set(record["Customer_ID"])

common_ids = client_ids.intersection(record_ids)

print("Client IDs:", len(client_ids))
print("Record IDs:", len(record_ids))
print("Common IDs:", len(common_ids))

print("Client IDs not in Record:", len(client_ids - record_ids))
print("Record IDs not in Client:", len(record_ids - client_ids))
```

```python
print("Churn counts:")
print(record["churn"].value_counts(dropna=False))

print("\nChurn percentage:")
print(round(record["churn"].value_counts(normalize=True, dropna=False) * 100, 2))
```

```python
df = client.merge(
    record,
    on="Customer_ID",
    how="inner",
    validate="one_to_one"
)

print("Merged dataframe shape:", df.shape)
display(df.head())
```

```python
print("Merged rows:", len(df))
print("Merged columns:", df.shape[1])

print("\nChurn distribution after merge:")
print(df["churn"].value_counts())
print(round(df["churn"].value_counts(normalize=True) * 100, 2))

print("\nFirst 10 columns:")
print(df.columns[:10].tolist())

print("\nLast 10 columns:")
print(df.columns[-10:].tolist())
```

```python
missing_count = df.isna().sum()
missing_pct = (missing_count / len(df)) * 100

missing_df = pd.DataFrame({
    "column": df.columns,
    "missing_count": missing_count.values,
    "missing_pct": missing_pct.values
}).sort_values("missing_pct", ascending=False)

display(missing_df.head(30))
```

```python
merged_path = TABLES_DIR / "merged_telecom_data_sample.csv"

df.head(1000).to_csv(merged_path, index=False)

print("Saved sample to:", merged_path)
```

```python
churn_counts = df["churn"].value_counts().sort_index()
churn_pct = df["churn"].value_counts(normalize=True).sort_index() * 100

plt.figure(figsize=(6, 4))
bars = plt.bar(["Not Churned", "Churned"], churn_counts.values)

plt.title("Customer Churn Distribution")
plt.ylabel("Number of Customers")

for i, value in enumerate(churn_counts.values):
    pct = churn_pct.iloc[i]
    plt.text(i, value, f"{value:,}\n({pct:.1f}%)", ha="center", va="bottom")

plt.tight_layout()

chart_path = CHARTS_DIR / "01_churn_distribution.png"
plt.savefig(chart_path, dpi=300, bbox_inches="tight")
plt.show()

print("Chart saved to:", chart_path)
```

```python
total_customers = len(df)
churned_customers = df["churn"].sum()
churn_rate = df["churn"].mean()

avg_monthly_revenue_all = df["rev_Mean"].mean()
avg_monthly_revenue_churned = df.loc[df["churn"] == 1, "rev_Mean"].mean()
avg_monthly_revenue_not_churned = df.loc[df["churn"] == 0, "rev_Mean"].mean()

monthly_revenue_at_risk = df.loc[df["churn"] == 1, "rev_Mean"].sum()
annual_revenue_at_risk = monthly_revenue_at_risk * 12

print("Total customers:", f"{total_customers:,}")
print("Churned customers:", f"{churned_customers:,}")
print("Churn rate:", f"{churn_rate:.2%}")

print("\nAverage monthly revenue - all customers:", round(avg_monthly_revenue_all, 2))
print("Average monthly revenue - churned customers:", round(avg_monthly_revenue_churned, 2))
print("Average monthly revenue - not churned customers:", round(avg_monthly_revenue_not_churned, 2))

print("\nMonthly revenue at risk:", round(monthly_revenue_at_risk, 2))
print("Annualized revenue at risk:", round(annual_revenue_at_risk, 2))
```
